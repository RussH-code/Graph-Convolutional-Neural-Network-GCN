# Graph Convolutional Neural Network (GCN)

```python
import numpy as np
import tensorflow as tf

## Spektral allows us to conveniently load many pre-processed datasets 
import spektral   ## Version 0.6.1
import scipy
```

## 1. Cora dataset

- adj - adjacency matrix
- features - feature vector
- labels - topic of eac paper
- mask - which node is in the training, validation and test set

```python
adj, features, labels, train_mask, val_mask, test_mask = spektral.datasets.citation.load_data(dataset_name="cora")
```

Spektral load data in sparse matrix form, here we will convert them to dense matrix form. Also, the adjacency matrix does not come with self edges, so we have to add that in manually. 

```python
adj = adj.todense() + np.eye(adj.shape[0]) ## Add an identity matrix the size of the adjacency matrix -- adding self edges
features = features.todense()

features = features.astype('float32')
adj = adj.astype('float32')
```

```python
labels.shape
```

```python
print(f'Training set: {np.sum(train_mask)}')
print(f'Validation set: {np.sum(val_mask)}')
print(f'Testing set: {np.sum(test_mask)}')
```

So the `cora` dataset has 2708 nodes and 1433 features. They come from 7 classes (topics). The dataset has 140 training, 500 validation and 1000 testing sets

For classification problem, we usually use cross entropy loss. Here we modify the loss function to accomodate masking of data sets.

## 2. Implementing graph neural network from Scratch

```python
def masked_softmax_cross_entropy(logits, labels, mask):
    loss = tf.nn.softmax_cross_entropy_with_logits(logits=logits, labels=labels)  ## Compute the loss
    mask = tf.cast(mask, dtype=tf.float32)   
    mask /= tf.reduce_mean(mask)     ## The loss function only applies to nodes in intended set
    loss *= mask                    
    return tf.reduce_mean(loss)

def masked_accuracy(logits, labels, mask):
    correct_prediction = tf.equal(tf.argmax(logits, 1), tf.argmax(labels, 1)) ## Check if logits predict the correct label
    accuracy_all = tf.cast(correct_prediction, tf.float32)
    mask = tf.cast(mask, dtype=tf.float32)
    mask /= tf.reduce_mean(mask)
    accuracy_all *= mask
    return tf.reduce_mean(accuracy_all)
```

### In each layer of the neural network, we aggregate information from the neighbour and represent that as a vector. This essentially helps us to learn relationships in a graph. The propagation rule:

### $ {H}^{l+1} = \sigma({W}\hat{A}{H}^{l}) $  

```python
def gnn(fts, adj, transform, activation):
    seq_fts = transform(fts)  ## Linear transformation with weights (W), the learnable parameters in neural network (basically pass data through neural network)
    ret_fts = adj @ seq_fts   ## Magic of GNN - matrix multiplication with adjacency matrix
    return activation(ret_fts)
```

```python
def train_cora(fts, adj, gnn_fn, units, epochs, lr):
    layer_1 = tf.keras.layers.Dense(units)  ## Hidden layer
    layer_2 = tf.keras.layers.Dense(7)      ## Classification - 7 outputs for 7 classes
    
    def cora_gnn(fts, adj):
        hidden = gnn_fn(fts, adj, layer_1, tf.nn.relu)
        logits = gnn_fn(hidden, adj, layer_2, tf.identity)
        return logits
    
    optimizer = tf.keras.optimizers.Adam(learning_rate=lr)
    
    best_accuracy = 0.0
    for ep in range(epochs + 1):
        with tf.GradientTape() as t:
            logits = cora_gnn(fts, adj)
            loss = masked_softmax_cross_entropy(logits, labels, train_mask)
            
        variables = t.watched_variables()
        grads = t.gradient(loss, variables)
        optimizer.apply_gradients(zip(grads, variables))
        
        logits = cora_gnn(fts, adj)
        val_accuracy = masked_accuracy(logits, labels, val_mask)
        test_accuracy = masked_accuracy(logits, labels, test_mask)
        
        if val_accuracy > best_accuracy:
            best_accuracy = val_accuracy
            print(f'Epoch: {ep}, Training loss: {loss.numpy()}, Val accuracy: {val_accuracy.numpy()}, Test accuracy: {test_accuracy.numpy()}')
```

```python
train_cora(features, adj, gnn, 32, 200, 1e-2)
```

## Baseline 

We can use an identity matrix as our adjacency matrix, this means the model won't be able to get information from neighbours. Basically, the model will only use information in its feature vector to classify nodes, but will not exploit relationships in graphs.

```python
train_cora(features, tf.eye(adj.shape[0]), gnn, 32, 200, 1e-2)
```

Unsurprisingly, we get a lower accuracy ~ 50%, adding relationships information improved the model by over 20%.

## Averaging

We may face diminishing or exploding gradient problem if the range of input is large. One way to normalize the adjacency matrix is to scale the adjacency matrix with the inverse of its diagonal node degree matrix (D). Essentially, we divide the features by the number of degrees (neighbours) at each node.

### $ \hat{A}{D}^{-1} $  

```python
D = np.zeros_like(adj)
np.fill_diagonal(D, np.sum(adj, axis=1))
D
```

```python
## Take the inverse of D

D_inverted = np.linalg.inv(D)
D_inverted
```

```python
A_averaged = adj @ D_inverted
```

```python
A_averaged
```

```python
train_cora(features, A_averaged, gnn, 32, 200, 1e-2)
```

There is a small but obvious improvement over using the un-normalized adjacency matrix. The model is able to achieve over 80% accuracy.

## Normalized laplacian matrix

In the <a href="https://arxiv.org/abs/1609.02907">original paper</a> describing the implementation of GCN by Thomas N. Kipf and Max Welling, they recommended a normalized version of the adjacency matrix as follows:

## $ \hat{A} = \tilde{D}^{-\frac{1}{2}} \tilde{A} \tilde{D}^{-\frac{1}{2}} $   where $\tilde{A} = {A} + {I}$ 

```python
D_invroot = np.linalg.inv(scipy.linalg.sqrtm(D)) 
A_hat = D_invroot @ adj @ D_invroot
```

```python
train_cora(features, A_hat, gnn, 32, 200, 1e-2)
```

## 3. Using Stellargraph

The StellarGraph library supports many state-of-the-art machine learning (ML) algorithms on graphs. The core of the GCN neural network model is a “graph convolution” layer. This layer is similar to a conventional dense layer, augmented by the graph adjacency matrix to use information about a node’s connections. It provides functions that we can use straight out of the box, so we don't have to build our GCN everytime.

```python
import pandas as pd
import os
import stellargraph as sg

from tensorflow import keras
import matplotlib.pyplot as plt
import matplotlib
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelBinarizer
```

The `Cora` dataset is in the stellargraph database as well.

```python
dataset = sg.datasets.Cora()
dataset.description
```

```python
G, node_subjects = dataset.load()
```

```python
print(G.info())
```

`G.info()` tells us some details about the data, such as the number of nodes, and the number of features.

```python
pd.DataFrame(node_subjects.value_counts())
```

`node_subjects` contains the labels, i.e. the subject a paper belongs to.

We can use machine learning style data processing with data from stellagraph.

```python
train_subjects, test_subjects = train_test_split(node_subjects, train_size=140, test_size=None, stratify=node_subjects)
val_subjects, test_subjects = train_test_split(test_subjects, train_size=500, test_size=None, stratify=test_subjects)
```

```python
matplotlib.style.use('fivethirtyeight')

subject_stratification1 = pd.DataFrame({"Training":train_subjects.value_counts()/len(train_subjects),
                                        "Validation": val_subjects.value_counts()/len(val_subjects),
                                        "Overall":node_subjects.value_counts()/len(node_subjects)})

subject_stratification1.plot.bar()
plt.ylabel("Percentage (%)")
plt.xticks(rotation=70)
plt.title("Class distribution for 7 subjects")
plt.show()
```

The class distribution is skewed towards neural networks, with stratification we ensure each class is equally represented in both training, validation and testing sets

```python
target_encoding = LabelBinarizer()

train_targets = target_encoding.fit_transform(train_subjects)
val_targets = target_encoding.transform(val_subjects)
test_targets = target_encoding.transform(test_subjects)
```

The CORA dataset contains attributes w_x that correspond to words found in that publication. If a word occurs more than once in a publication the relevant attribute will be set to one, otherwise it will be zero. These numeric attributes have been automatically included in the StellarGraph instance G, and so we do not have to do any further conversion.

## Creating GCN layers

A machine learning model in Stellargraph consists of a pair of items:

1. The layers - graph convolution layers, dropout, pooling or even dense layers
2. Data generator - convert the graph structure and node features into a format that can be fed into keras model for training and prediction. For this we will use the `FullBatchNodeGenerator` from `stellargraph`.

```python
## Specifying the method='gcn' argument will yield data appropriate for the GCN algorithm specifically
## by using the normalized graph Laplacian matrix to capture the graph structure.

generator = sg.mapper.FullBatchNodeGenerator(G, method="gcn")
```

A generator encodes the data required to produce the model inputs. Calling the `flow` method with a set of nodes and their true labels will produce an object that can be 

```python
train_gen = generator.flow(train_subjects.index, train_targets)
```

To create GCN layers, we can use the built-in GCN constructor from `stellargraph.layer`. Four arguments need to be specified:

1. `layer_sizes`: the number of hidden GCN layer and the number of nodes as a list 
2. `activations`: activation function foe each layer
3. `generator`: data generator that passes in the input data
4. `dropout`: the dropout rate at each GCN layer

```python
gcn = sg.layer.GCN(layer_sizes=[16, 16],         ## 2 layers with 16 nodes each
                   activations=["relu", "relu"], 
                   generator=generator, 
                   dropout=0.5)
```

To create a `Keras` model, we now expose the input and output layer of the GCN layer with the `GCN.in_out_tensors()` method

```python
x_inp, x_out = gcn.in_out_tensors()

x_inp
```

```python
x_out
```

The output of the GCN layer is a 16-dimensional vector. Since `stellargraph` is built on `Keras`, to perform class predictions, we can simply add a Dense layer with softmax activation using standard `Keras` functionality. This ensures the final output vector has the right number of dimensions (one for each class) and the output will be probabilities that sum to 1.

```python
predictions = keras.layers.Dense(units=train_targets.shape[1], activation="softmax")(x_out)
```

# Training model and evaluation

Now we can create the actual model with input as `x_inp` and output as `predictions`. Just like other neural network model, we have to compile it and specify optimizers, loss and metrics.

```python
model = keras.Model(inputs=x_inp, outputs=predictions)
model.compile(
    optimizer=keras.optimizers.Adam(lr=0.01),
    loss=keras.losses.categorical_crossentropy,
    metrics=["acc"],
)
```

When training the model, we also want to look at at validation metrics, so we create a data generator for validation set as well. Moreover, we also want to implement early stopping.

```python
val_gen = generator.flow(val_subjects.index, val_targets)
```

```python
es_callback = keras.callbacks.EarlyStopping(monitor="val_acc", patience=50, restore_best_weights=True)
```

```python
history = model.fit(
    train_gen,
    epochs=200,
    validation_data=val_gen,
    verbose=2,
    shuffle=False,  # this should be False, since shuffling data means shuffling the whole graph
    callbacks=[es_callback],
)
```

```python
sg.utils.plot_history(history)
```

We can also test our models against the test set. We first create a data generator for the test set.

```python
test_gen = generator.flow(test_subjects.index, test_targets)
```

```python
model.evaluate(test_gen)
```

## Making preictions for all nodes

```python
all_nodes = node_subjects.index
all_gen = generator.flow(all_nodes)
all_predictions = model.predict(all_gen)
```

```python
all_predictions
```

The predictions generated by the model are the probabilities of the node being in each class. We can use `inverse_transform` method of `labelBinarizer` to return the target categories.

```python
node_predictions = target_encoding.inverse_transform(all_predictions.squeeze())
```

```python
df = pd.DataFrame({"Predicted": node_predictions, "Ground Truth": node_subjects})
df.head(10)
```

## Node embeddings

Node embedding algorithms compute low-dimensional vector representations of nodes in a graph, which will give us insight into nodes’ similarity and network structure. Given the widespread prevalence of graphs on the web and in the physical world, representation learning on graphs plays a significant role in a wide range of applications, such as link prediction and anomaly detection. 

We can use the last hidden graph convolution layer of the GCN model since the outputs are lower-dimensionality representations of the original data. We do not want to train this model on the labels. The idea is after passing through the convolution layers, each node would have gathered some information about its neighbours. The output of the model should output similar values for nodes that are similar (hopefully).

```python
embedding_model = keras.Model(inputs=x_inp, outputs=x_out)
emb = embedding_model.predict(all_gen)
emb.shape
```

The result still has 16 dimensions, which is too much to plot. We can further use dimensionality reductions such as PCA, T-SNE.

```python
from sklearn.manifold import TSNE
```

```python
def dim_reduction_plot(transform, data):
    X = data.squeeze()
    trans = transform(n_components=2)
    X_reduced = trans.fit_transform(X)

    fig, ax = plt.subplots(figsize=(7, 7))

    ax.scatter(X_reduced[:, 0],
              X_reduced[:, 1],
              c=node_subjects.astype("category").cat.codes,
              cmap="jet",
              alpha=0.7)

    ax.set_title(f"{transform.__name__} visualization of GCN embeddings for cora dataset")
    ax.set_xlabel("$X_1$")
    ax.set_ylabel("$X_2$")
```

```python
dim_reduction_plot(TSNE, emb)
```

We can see that the data can be largely seperated into 7 clusters, the orange cluster might be more difficult to identify as it's less distinct from the others.
